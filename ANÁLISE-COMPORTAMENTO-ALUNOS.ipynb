```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('ALUNOS-ACESSOS-TODOS.csv')
```

```python
data['TS_TIME'] = pd.to_datetime(data['TS_TIME'])
```

```python
daily_access = data.groupby(data['TS_TIME'].dt.date).size()
```

```python
sns.set(style="darkgrid")
plt.style.use("dark_background")

plt.figure(figsize=(14, 7))
sns.lineplot(x=daily_access.index, y=daily_access.values, marker='o', color='skyblue')
plt.title('Tendência de Acessos Diários ao Site', fontsize=16)
plt.xlabel('Data', fontsize=14)
plt.ylabel('Número de Acessos', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_access = data.groupby(data['TS_TIME'].dt.to_period('M')).size()
```

```python
sns.set(style="darkgrid")
plt.style.use("dark_background")

plt.figure(figsize=(14, 7))
sns.barplot(x=monthly_access.index.astype(str), y=monthly_access.values, palette="Blues_r")
plt.title('Tendência de Acessos Mensais ao Site', fontsize=16)
plt.xlabel('Mês', fontsize=14)
plt.ylabel('Número de Acessos', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

daily_percentage = (daily_access / daily_access.sum()) * 100
```

```python
sns.set(style="darkgrid")
plt.style.use("dark_background")

plt.figure(figsize=(14, 7))
sns.lineplot(x=daily_percentage.index, y=daily_percentage.values, marker='o', color='skyblue')
plt.title('Porcentagem de Acessos Diários ao Site', fontsize=16)
plt.xlabel('Data', fontsize=14)
plt.ylabel('Porcentagem de Acessos (%)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_access_pct_change = monthly_access.pct_change() * 100

plt.figure(figsize=(14, 7))
sns.barplot(x=monthly_access_pct_change.index.astype(str), y=monthly_access_pct_change.values, palette="Blues_r")
plt.title('Variação Percentual Mês a Mês nos Acessos', fontsize=16)
plt.xlabel('Mês', fontsize=14)
plt.ylabel('Variação Percentual (%)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

monthly_percentage = (monthly_access / monthly_access.sum()) * 100
```

```python
sns.set(style="darkgrid")
plt.style.use("dark_background")

plt.figure(figsize=(14, 7))
sns.barplot(x=monthly_percentage.index.astype(str), y=monthly_percentage.values, palette="Blues_r")
plt.title('Porcentagem de Acessos Mensais ao Site', fontsize=16)
plt.xlabel('Mês', fontsize=14)
plt.ylabel('Porcentagem de Acessos (%)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data['ORIGIN'].isnull().sum()  
```

```python
origin_counts = data['ORIGIN'].value_counts()
```

```python
origin_percentage = (origin_counts / origin_counts.sum()) * 100
```

```python
sns.set(style="darkgrid")
plt.style.use("dark_background")

plt.figure(figsize=(10, 6))
sns.barplot(x=origin_percentage.index, y=origin_percentage.values, palette="Blues_r")
plt.title('Distribuição de Acessos por Origem (%)', fontsize=16)
plt.xlabel('Origem', fontsize=14)
plt.ylabel('Porcentagem de Acessos (%)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
access_count_per_user = data['EMAIL'].value_counts()
```

```python
sns.set(style="darkgrid")
plt.style.use("dark_background")

plt.figure(figsize=(14, 7))
access_count_distribution = access_count_per_user.value_counts().sort_index()
sns.barplot(x=access_count_distribution.index, y=access_count_distribution.values, palette="Blues_r")
plt.title('Frequência de Acessos por Usuário', fontsize=16)
plt.xlabel('Número de Acessos', fontsize=14)
plt.ylabel('Número de Usuários', fontsize=14)
plt.xticks(rotation=45)
plt.xlim(0, 30)  # Limita o eixo X a 50 acessos
plt.ylim(0, 5000)  # Limita o eixo Y a 5000 usuários
plt.tight_layout()
plt.show()
```
